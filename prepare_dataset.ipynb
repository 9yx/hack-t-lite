Создание `Goals` внутри команд и категорий (асситентов):

```python
def goal_prompt(command):
    prompt = {
        "modelUri": f"{yandex_gtp_url}",
        "completionOptions": {
            "stream": False,
            "temperature": 0.2,
            "maxTokens": "2000"
        },
        "messages": [
            {
                "role": "system",
                "text": """Use english only. To generate from 6 to 30 GOALS for a given Command, consider the following: 1. Identify the primary purpose of the Command. 2. Break down the Command's functionality into key components. 3. Determine the desired outcomes or results of using the Command. 4. Consider potential user needs or expectations related to the Command. 5. Formulate clear, actionable, and measurable GOALS that align with the Command's purpose and functionality. 6. GOALS contains example of parameters values from the command args Example Command: "Create Budget: \"create_budget\", args: \"income\": \"<monthly_income>\", \"expenses\": \"<list_of_expenses>\"", For example, if the Command is "create_budget", potential one GOALS could be: "1. Accurately input and categorize all income sources. 2. List and categorize all monthly expenses comprehensively. 3. Allocate funds for savings and emergency fund. 4. Monthly income is $1000 5. Generate a balanced budget that aligns with financial objectives. 6. ... 7. ... 8. ... 9. ... 10. ... ... " Ensure that the GOALS are specific, relevant to the Command, and contribute to its effective utilization."""
            },
            {
                "role": "user",
                "text": f"Command: \"{command}\""
            }
        ]
    }
    return prompt

def create_goals():
    with open("./commands/commands.json", "r") as json_file:
        loaded_dict = json.load(json_file)
        for val in loaded_dict[10:]:
            category = val['category']
            commands = val['commands']
            print(f"Start {category}")
            result_goals = []
            for command in commands:
               try:
                    print(f"Start {command}")
                    prompt = goal_prompt(command)
                    time.sleep(3)
                    response = yandex_gpt(prompt)
                    lst = response['result']['alternatives'][0]['message']['text'].split('\n')[1:]
                    lst = [re.sub(r'^\d+\.\s*', '', item).strip() for item in lst]
                    result_goals.append({"command": command, "goals": lst})
                    print(f"End {command}, goals: {len(lst)}")
               except:
                   print(f"Error: {response}")
            with open(f"goals/{category}.json", "w") as goal_json_file:
                print(f"Save {category}")
                json.dump(result_goals, goal_json_file, indent=4)
        
```

Создание промтов (подобных как в были в описании кейса), по которым получим JSON ответов (они промежуточные, так тут нет выбора в командах, только возможность выбрать правильный, чтобы YandexGPT не ошибся).

```python
def create_goal_samples(data, num_samples, min_length=3, max_length=5, max_attempts=1000):
    unique_samples = set()
    data_length = len(data)
    attempts = 0
    
    while len(unique_samples) < num_samples and attempts < max_attempts:
        if data_length == 1:
            sample = tuple(data)
        else:
            length = min(random.randint(min_length, max_length), data_length)
            sample = tuple(sorted(random.sample(data, length)))
        
        unique_samples.add(sample)
        attempts += 1
    
    if attempts >= max_attempts:
        print(f"Warning: Maximum attempts ({max_attempts}) reached. Only {len(unique_samples)} unique samples were generated.")
    
    return list(unique_samples)

def remove_extra_spaces(text):
    # Удаляем пробелы в начале и конце строки
    text = text.strip()
    # Разбиваем строку на слова и соединяем их обратно с одним пробелом
    return ' '.join(text.split())


def create_result_prompt(category, goals, commands):
    descr = " ".join(descriptions[category])
    indexed_goals = "".join([f"{i + 1}. {s}\n" for i, s in enumerate(goals)])
    indexed_commands = "".join([f"{i + 1}. {s}\n" for i, s in enumerate(commands)])

    prompt = {
        "modelUri": f"{yandex_gtp_url}",
        "completionOptions": {
            "stream": False,
            "temperature": 0.3,
            "maxTokens": "2000"
        },
        "messages": [
            {
                "role": "system",
                "text": f"Use english only. You are {category}, {descr}. Ensure the response can be parsed by Python json.loads. Return only json."
            },
            {
                "role": "user",
                "text": remove_extra_spaces(f"""
                 GOALS:\n""" + indexed_goals + """\n

                 Constraints:
                 1. ~4000 word limit for short term memory. Your short term memory is short, so immediately save important information to files.
                 2. If you are unsure how you previously did something or want to recall past events, thinking about similar events will help you remember.
                 3. No user assistance
                 4. Exclusively use the commands listed in double quotes e.g. "command name"
                 5. Use subprocesses for commands that will not terminate within a few minutes

                 Commands:\n""" + indexed_commands + """\n

                 Resources:
                 1. Internet access for searches and information gathering.
                 2. Long Term memory management.
                 3. GPT-3.5 powered Agents for delegation of simple tasks.
                 4. File output.

                 Performance Evaluation:
                 1. Continuously review and analyze your actions to ensure you are performing to the best of your abilities.
                 2. Constructively self-criticize your big-picture behavior constantly.
                 3. Reflect on past decisions and strategies to refine your approach.
                 4. Every command has a cost, so be smart and efficient.

                 Aim to complete tasks in the least number of steps. You should only respond in JSON format as described below Response Format: { "thoughts": { "text": "thought", "reasoning": "reasoning", "plan": "- short bulleted\n- list that conveys\n- long-term plan", "criticism": "constructive self-criticism", "speak": "thoughts summary to say to user" }, "command": { "name": "command name", "args": { "arg name": "value" } } }"
            """)
            }
        ]
    }
    return prompt


def find_json_object(s):
    braces = 0
    for i, c in enumerate(s):
        if c == '{':
            braces += 1
        elif c == '}':
            braces -= 1
            if braces == 0:
                return s[:i + 1]
    return None


def create_prompt_for_teach():
    count = 0
    for filename in os.listdir('goals'):
        if filename.endswith('.json'):
            file_path = os.path.join('goals', filename)
            with open(file_path, 'r') as file:
                data = json.load(file)
                category = os.path.splitext(filename)[0]
                print(f"Process category: {category}")
                result_prompts = []
                for command_data in data:
                    command = command_data['command']
                    goals = command_data['goals']
                    goals_samples = create_goal_samples(goals, 10)

                    for goal_list in goals_samples:
                        print(f"Process category: {category}, file: {count}")
                        prompt = create_result_prompt(category, goal_list, [command])
                        time.sleep(3)
                        response = yandex_gpt(prompt)
                        try:
                            cleaned_text = (response['result']['alternatives'][0]['message']['text']
                                            .strip()
                                            .replace("*", "")
                                            .replace("json", "")
                                            .replace("`", "")
                                            )
                            json_obj = json.loads(cleaned_text)
                            result_prompts.append({"category": category, "commands": [command], "goals": goal_list, "result": json_obj})
                        except Exception as e:
                            print(f"Error {e} in response: |{cleaned_text}|")
                        count += 1
                with open(f"prompts/{category}.json", "w") as prompts_json_file:
                    json.dump(result_prompts, prompts_json_file, indent=4)
    print(count)
```
